```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
import nltk
```

```python
dataset_df = pd.read_csv('./php_xss_datasets.csv', encoding='utf-8')
opcode_list = np.array(dataset_df['opcode'].tolist())
print(opcode_list)
```

```python
corpus = []
for opcode in opcode_list:
    corpus.append(nltk.word_tokenize(opcode))
corpus = np.array(corpus)
w2v_model = Word2Vec(corpus, size=128, window=5, min_count=1)
```

```python
print(corpus[0])
```

```python
w2v_model.save('opcode.word2vec')
```

```python
wv = np.array(w2v_model.wv.vectors)
words = np.array(w2v_model.wv.index2word)
print(wv.shape)
print(words.shape)
word_dict = dict(zip(words, wv))
```
