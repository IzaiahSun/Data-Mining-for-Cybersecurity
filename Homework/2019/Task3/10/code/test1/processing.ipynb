```python
from utils import GeneSeg
import csv,pickle,random,json
import numpy as np
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
import time
import pandas as pd
```

```python

vec_dir="data\\word2vec.pickle"

process_datas_dir="data\\process_datas.pickle"
```

```python
with open(vec_dir,"rb") as f :
    word2vec=pickle.load(f)
    dictionary=word2vec["dictionary"]
    reverse_dictionary=word2vec["reverse_dictionary"]
    embeddings=word2vec["embeddings"]
```

```python
xssed_data=[]
normal_data=[]
```

```python
with open("data\\xssed.csv","r",encoding="utf-8") as f:
    reader = csv.DictReader(f, fieldnames=["payload"])
    for row in reader:
        payload=row["payload"]
        word=GeneSeg(payload)
        xssed_data.append(word)

with open("data\\dmzo_nomal.csv","r",encoding="utf-8") as f:
    reader=csv.reader(f)
    reader = csv.DictReader(f, fieldnames=["payload"])
    for row in reader:
        payload=row["payload"]
        word=GeneSeg(payload)
        normal_data.append(word)
```

```python
xssed_data[-1]
```

```python

xssed_num=len(xssed_data)
normal_num=len(normal_data)

xssed_labels=[1]*xssed_num  #1代表XSS
normal_labels=[0]*normal_num #0代表正常

datas=xssed_data+normal_data
labels=xssed_labels+normal_labels
```

```python
labels=to_categorical(labels)
```

```python
labels[:10]
```

```python
def to_index(data):
    d_index=[]
    for word in data:
        if word in dictionary.keys():
            d_index.append(dictionary[word])
        else:
            d_index.append(dictionary["UNK"])
    return d_index
```

```python
datas_index=[to_index(data) for data in datas]
```

```python
datas_index[:10]
```

```python
datas_index=pad_sequences(datas_index,value=-1)
```

```python
print(len(datas_index[0]))
datas_index[0]
```

```python
rand=random.sample(range(len(datas_index)),len(datas_index))
```

```python
print(len(rand))
rand[:10]
```

```python
datas_index[0].tolist()+labels[0].tolist()
```

```python
#交换数据的顺序：原数据正常语句或xss语句太过集中
datas=[datas_index[index].tolist()+labels[index].tolist()  for index in rand]
```

```python
str(datas[0]).replace('[','').replace(']','')
```

```python

pre_datas="data\\pre_datas.csv"

dims_num = embeddings["UNK"].shape[0]
word2vec["dims_num"]=dims_num
with open(vec_dir,"wb") as f :
    pickle.dump(word2vec,f)
print("Saved word2vec to:",vec_dir)


print("Write datas to:",pre_datas)   
with open(pre_datas,"w") as f:
    for i in range(len(datas)):
        data_line=str(datas[i]).replace('[','').replace(']','')+"\n"
        f.write(data_line)


```
