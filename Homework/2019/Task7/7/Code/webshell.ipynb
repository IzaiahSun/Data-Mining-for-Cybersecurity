```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import os
import re
import subprocess
%matplotlib inline
import warnings
 
warnings.filterwarnings('ignore')
```

```python
def load_file(filename):
    x = ''
    with open(filename, encoding='utf-8',errors='ignore') as f:
        for line in f:
            line = line.strip('\r')
            x += line
    return x
```

```python
def load_files_re(dir):
    files_list = []
    g = os.walk(dir)
    for path, d, filelist in g:
        for filename in filelist:
            if filename.endswith('.php'):
                fulepath = os.path.join(path, filename)
#                 fulepath = str(path) + '/' + str(filename)
                print("Load %s" % fulepath)
#                 t= load_file(fulepath)
                files_list.append(fulepath)
    return files_list
```

```python
webshell_list = load_files_re('php_webshell')
```

```python
normal_list = load_files_re('white')
```

```python
labels_webshell = []
labels_common = []
for i in range(0,len(webshell_list)):
    labels_webshell.append(1)
for i in range(0,len(normal_list)):
    labels_common.append(0)
```

```python
file = webshell_list + normal_list
labels = labels_webshell + labels_common
```

```python
datadict = {'label':labels,'file':file}
df = pd.DataFrame(datadict,columns=['label','file'])
```

```python
df['file'] = df['file'].map(lambda x: os.getcwd() + '\\' + x)
```

```python
def getopcode(x):
    try:
        cmd = "php -dvld.active=1 -dvld.execute=0 \"" + str(x) + '\"'
        print(cmd)
        output = subprocess.getoutput(r''+cmd)
        oplist = re.findall(r'\s(\b[A-Z_]+\b)\s',output)
        return oplist
    except:
        print("error " + str(x))
        return None

df['opcode'] = df['file'].map(lambda x:getopcode(x))
```

```python
df = df.dropna()
def getoplen(x):
    return len(x) 
df['opcode_len'] = df['opcode'].map(lambda x:getoplen(x))
```

```python
from gensim.models import Word2Vec
def getw2v(opc_list,label_list):
    print(label_list[0:10])
    stop = []
    w2v_list = []
    for i in range(0,3597):
        try:
            print(label_list[0:10])
            tmp = []
            name = opc_list[i]
            for j in range(0,len(name)):
                tmp.append(name[j])
            w2v_list.append(tmp)
        except:
            pass
 
    model = Word2Vec(w2v_list, min_count = 1)
    model.wv.save_word2vec_format('word2vec.txt',binary=False) 
    label_vect = []
    wv_vect = []
    for i in range(0,3597):
        try:
            #print(i)
            name = opc_list[i]
            tmp = []
            vect = []
            for j in range(0,len(name)):
                if name[j] in stop:
                    continue
                tmp.append(model[name[j]])
                if j >= 99:
                    break
            if len(tmp) < 100:
                for k in range(0,100-len(tmp)):
                    tmp.append([0]*100)
            vect = np.vstack((x for x in tmp))
            wv_vect.append(vect)
            label_vect.append(label_list[i])
        except:
            pass
    wv_vect = np.array(wv_vect)
    label_vect = np.array(label_vect)
    return wv_vect,label_vect
 
w2v_word_list,label_list = getw2v(df['opcode'],df['label'])
```

```python
X = w2v_word_list
Y = label_list
```

```python
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense,Embedding
from keras.layers import LSTM
 
model = Sequential()
model.add(LSTM(128,dropout = 0.2,recurrent_dropout = 0.2))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=2)
model.fit(X, Y, validation_split = 0.2, epochs = 20, batch_size = 100, verbose = 1, callbacks= [early_stopping])
```

```python
import gensim.models
def word2vec(opcode):
    model = gensim.models.KeyedVectors.load_word2vec_format('word2vec.txt', binary=False)
    tmp = []
    stop = []
    wv_vect = []
    for j in range(0,len(opcode)):
        if opcode[j] in stop:
            continue
        tmp.append(model[opcode[j]])
        if j >= 99:
            break
    if len(tmp) < 100:
        for k in range(0,100-len(tmp)):
            tmp.append([0]*100)
    vect = np.vstack((x for x in tmp))
    wv_vect.append(vect)
    wv_vect = np.array(wv_vect)
    return wv_vect
```

```python
def test(filepath):
    opcode = getopcode(filepath)
    vec = word2vec(opcode)
    pre = model.predict(vec)
    print(pre)
    print("webshell") if pre > 0.5 else print("Normal")
```
