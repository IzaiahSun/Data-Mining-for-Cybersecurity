```python
import pandas as pd
import numpy as np
from math import log
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data.csv', encoding='utf-8')

def process_label(x):
    if x == 'bad':
        return 0
    elif x == "good":
        return 1
    else:
        print("???")

data['label'] = data['label'].map(lambda x: process_label(x))
```

```python
import urllib.parse
import tldextract
import whois
import re
```

```python
data['url_length'] = data['url'].map(lambda x: len(x))
data['count_slant'] = data['url'].map(lambda x: x.count('/'))
    
data['url'] = data['url'].map(lambda x: 'http://' + x if re.match(r'^(?:http|ftp)s?://', x) is None else x)

data
```

```python
data['subdomain'] = data['url'].map(lambda x: tldextract.extract(x)[0])
data['domain'] = data['url'].map(lambda x: tldextract.extract(x)[1])
data['suffix'] = data['url'].map(lambda x: tldextract.extract(x)[2])
```

```python
from unicodedata import normalize
data['url'] = data['url'].map(lambda x: normalize('NFKD',x))
data['path'] = data['url'].map(lambda x: urllib.parse.urlparse(x)[2])
data['query'] = data['url'].map(lambda x: urllib.parse.urlparse(x)[4])
```

```python
data['count_dot'] = data['url'].map(lambda x: x.count('.'))
data['path_len'] = data['path'].map(lambda x: len(x))
data['query_len'] = data['query'].map(lambda x: len(x))
```

```python
data
```

```python
def process(x):
    if x != '':
        return 1
    else:
        return -1
```

```python
data['subdomain'] = data['subdomain'].map(lambda x: process(x))
data['path'] = data['path'].map(lambda x: process(x))
data['query'] = data['query'].map(lambda x: process(x))
```

```python
data['domain_len'] = data['domain'].map(lambda x: len(x))
data['suffix_len'] = data['suffix'].map(lambda x: len(x))
data = data.drop('url', axis=1)
data = data.drop('domain', axis=1)
data = data.drop('suffix', axis=1)

data
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
x,y = data.ix[:,1:],data.ix[:,0]
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
#random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
Y_pred = random_forest.predict(X_test)
accuracy_score(Y_test, Y_pred)
```

```python
precision_score(Y_test, Y_pred)
```

```python
recall_score(Y_test, Y_pred)
```

```python
def get_url_length(url):
    return len(str(url))

def get_url_slant_num(url):
    return url.count('/')

def has_subdomain(url):
    return 1 if tldextract.extract(url)[0] is not None else 0

def has_path(url):
    return 1 if urllib.parse.urlparse(url)[2] is not None else 0

def has_query(url):
    return 1 if urllib.parse.urlparse(url)[4] is not None else 0

def get_url_dot_num(url):
    return url.count('.')

def get_path_length(url):
    return len(urllib.parse.urlparse(url)[2])

def get_query_length(url):
    return len(urllib.parse.urlparse(url)[4])

def get_domain_lentgh(url):
    return len(tldextract.extract(url)[1])

def get_suffix_length(url):
    return len(tldextract.extract(url)[2])

def get_url_features(url):
    URL = 'http://' + url if re.match(r'^(?:http|ftp)s?://', url) is None else url
    return [get_url_length(url), get_url_slant_num(url), has_subdomain(URL), has_path(URL), has_query(URL), 
           get_url_dot_num(URL), get_path_length(URL), get_query_length(URL), get_domain_lentgh(URL), get_suffix_length(URL)]

def whether_good(test_url):
    test_features = []
    test_features.append(get_url_features(test_url))
    pre = random_forest.predict(test_features)
    print(1) if pre == 1 else print(0)
```
